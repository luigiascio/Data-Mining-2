```python
import pandas as pd
import requests
import time

# Inserisci la tua API key OMDb qui
OMDB_API_KEY = "da cambiare"

def get_title_from_imdb_id(imdb_id):
    url = f"http://www.omdbapi.com/?i={imdb_id}&apikey={OMDB_API_KEY}"
    response = requests.get(url)
    if response.status_code == 200:
        print('Ce semo!')
        data = response.json()
        return data.get("Title", None)
    return None


# Carica il CSV
df = pd.read_csv('id_movies.csv')

# Controlla che esista la colonna 'id'
if 'id' not in df.columns:
    raise ValueError("Il file CSV deve contenere una colonna chiamata 'id'.")

# Ottieni i titoli per ogni ID
titles = []
for imdb_id in df['id']:
    title = get_title_from_imdb_id(imdb_id)
    titles.append(title)
    time.sleep(0.1)  # Piccola pausa per evitare rate limit

# Aggiunge la colonna 'title' al DataFrame
df['title'] = titles

df.to_csv('bro2.csv', index=False)

print(f"File salvato")
```
